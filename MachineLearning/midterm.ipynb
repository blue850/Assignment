#2) 데이터셋 나누기

```python
import pandas as pd
```

```python
data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Machine Learning/data/MLatoz_F.csv', encoding='utf-8', index_col=0)
```

```python
X=data[['교육수준_고졸','교육수준_대학','교육수준_대학원','교육수준_중졸','연령','직장근무년수','거주지거주년수','수입','대출_수입비율','신용카드대출금','기타대출금']]
```

```python
y=data[['상환불이행여부']]
```

## 분석용 데이터와 신규고객데이터로 나누기

```python
analyze_data = data.loc[0:699] # analyze_data = 분석용 데이터
```

```python
analyze_data
```

```python
analyze_data_X = analyze_data.loc[:, '교육수준_고졸':'기타대출금']
```

```python
analyze_data_X
```

```python
analyze_data_y = analyze_data[['상환불이행여부']]
```

```python
analyze_data_y
```

```python
new_data=data.loc[700:] #new_data = 신규고객데이터
```

```python
new_data
```

```python
new_data_X = new_data.loc[:, '교육수준_고졸':'기타대출금']
```

```python
new_data_X
```

```python
new_data_y = new_data[['상환불이행여부']]
```

```python
new_data_y
```

## 분석용 데이터를 train-test 데이터셋으로 나누기

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(analyze_data_X, analyze_data_y, stratify=analyze_data_y, random_state=20, test_size=0.3)
```

```python
X_test.shape
```

#3) KNN 모델 적용 및 Random Search

###KNN 모델 적용

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

### Random Search

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
param_distribs = {'n_neighbors': randint(low=1, high=20)}
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
random_search = RandomizedSearchCV(KNeighborsClassifier(), param_distributions=param_distribs, cv=5, return_train_score=True )
```

```python
random_search.fit(X_train, y_train)
```

```python
print("Best parameter : {}".format(random_search.best_params_))
```

```python
print("Best Cross-validity Score : {:.3f}".format(random_search.best_score_))
```

```python
print("Test set Score : {:.3f}".format(random_search.score(X_test, y_test)))
```

```python
result_random = pd.DataFrame(random_search.cv_results_)
```

```python
result_random
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(result_random['param_n_neighbors'],result_random['mean_train_score'], label="Train")
plt.plot(result_random['param_n_neighbors'],result_random['mean_test_score'], label="Test")
plt.legend()
```

#4) 최종 KNN 모델 생성 및 결과 평가

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn12=KNeighborsClassifier(n_neighbors=12) #가장 적합한 파라미터인 12를 적용한 knn 모델
```

```python
knn12.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
pred_train=knn12.predict(X_train)
```

```python
confusion_train = confusion_matrix(y_train, pred_train)
```

```python
print("훈련데이터 오차행렬: \n",confusion_train)
```

```python
pred_test=knn12.predict(X_test)
```

```python
confusion_test=confusion_matrix(y_test, pred_test)
```

```python
print("테스트데이터 오차행렬: \n",confusion_test)
```

```python
from sklearn.metrics import f1_score
```

```python
TP=21
FP=9
FN=34
TN=146
```

```python
precision=TP/(TP+FP)
precision
```

```python
recall=TP/(TP+FN)
recall
```

```python
f_score=2/((1/precision)+(1/recall))
f_score
```

```python
f1=f1_score(y_test, pred_test)
```

```python
print("f1 score:\n", f1)
```

```python
from sklearn.metrics import classification_report
```

```python
cfreport=classification_report(y_test, pred_test)
```

```python
print("분류예측 레포트:\n", cfreport)
```
