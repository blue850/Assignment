##[문제 3] 통계 모델링에서 회귀 분석은 변수 간의 관계를 추정하기 위한 일련의 통계 프로세스이다. 여기에는 종속 변수와 하나 이상의 독립 변수(또는 '예측 변수') 간의 관계를 파악하고자 할때 여러 변수를 모델링하고 분석하는 많은 기술이 포함된다. 로지스틱 회귀의 정확성을 얻기 위한 Python 프로그램을 작성하시오. 결과로써 정확도, "Confusion matrix", "Classfication Report"를 보여주고, 결과를 시각화하시오(2점)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.datasets import load_iris
import seaborn as sns

feature = iris.data
target = iris.target

feature_train, feature_test, target_train, target_test = train_test_split(feature, target, test_size=0.3, random_state=42)

model = LogisticRegression() #로지스틱 회귀 모델생성
model.fit(feature_train, target_train) #훈련 데이터에 모델 적용

feature_pred = model.predict(feature_train)
confusion_train= confusion_matrix(target_train, feature_pred) #훈련데이터 오차행렬

target_pred = model.predict(feature_test)
confusion_test=confusion_matrix(target_test, target_pred) #테스트데이터 오차행렬

accuracy = accuracy_score(target_test, target_pred) #정확도 계산
print('정확도 :',accuracy)
print('\n')

print("Confusion matrix - train\n", confusion_train)
print("Confusion matrix - test\n", confusion_test)
print('\n')

report=classification_report(target_test, target_pred)
print("Classfication Report:\n", report)

sns.heatmap(confusion_test, annot=True, cmap='Blues') #Confusion Matrix를 히트맵으로 시각화
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion matrix')
plt.show()
```

#Logistic Regression
- 장점: 선형 회귀에 비해 이상치에 대해 덜 민감하게 반응함
- 단점: 복잡한 데이터에서는 적합한 모델을 도출하기 어려움

#Decision Tree
- 장점: 분석 결과가 트리 구조로 시각적으로 표현되기 때문에 쉽게 이해하고 설명할 수 있음
- 단점: 분류시 1개 변수만을 단계적으로 고려하기 때문에 분류 정확도가 다른 알고리즘에 비해 낮음, 특정 변수가 명확히 집단을 구분하는 데에 기여하지 못하면 분류율이 떨어지고, 트리 구조가 복잡해지는 문제 발생, 몇 개 케이스의 변경, 유사한 기여도를 보이는 변수의 비교에 따라 트리 구성이 크게 달라질 수 있음

#Random Forest
- 장점: 의사 결정 트리는 학습 데이터에 따라 트리 구조가 달라지기 때문에 일반화하여 사용하는 데 안정적이지 못한 반면, 랜덤포레스트는 일반화 성능을 향상시켜 정확도와 안정성이 좋음, 노이즈가 포함된 데이터, 데이터 분포가 고르지 않은 경우에도 적합함
- 단점: 연산 시간이 오래 걸림

#SVM
- 장점: 딥러닝 이전에 분류에 있어 매우 좋은 성능을 보였음, 커널함수를 설계하면 나머지는 자동으로 분류기가 계산하여 활용하기 용이함
- 단점: 데이터가 클수록 시간 소모가 매우 높음, 초평면의 최대 마진을 구하기 위해 훈련 데이터 개수의 제곱에 해당하는 계산량이 필요함, 모든 서포트 벡터를 저장해야 함
